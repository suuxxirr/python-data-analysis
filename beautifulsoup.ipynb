# 기상청 웹페이지에서 도시별 현재 기온과 습도 정보 추출

[대상 사이트 주소](https://www.weather.go.kr/w/weather/land/city-obs.do)



```python
import requests
from bs4 import BeautifulSoup

response = requests.get('https://www.weather.go.kr/w/weather/land/city-obs.do')
soup = BeautifulSoup(response.text, 'html.parser')

table = soup.find('table', {'class':'table-col'})
data =[]
for tr in table.find_all('tr'):
  tds = list(tr.find_all('td'))

  for td in tds:
    if td.find('a'): # <td> 안에 <a> 태그가 있으면
      city = td.find('a').text
      temperature = tds[5].text
      humidity = tds[10].text
      data.append([city, temperature, humidity])

data # 주피터 노트북에서는 print를 사용하지 않아도 변수의 값 표시됨
```

# 추출한 데이터를 CSV 파일에 저장하기

```python
with open('weather.csv', 'w') as file: # weather.csv 파일에 write 모드로 작성
  file.write('city,temperature,humidity\n') # 머리글
  for i in data:
    file.write(f'{i[0]},{i[1]},{i[2]}\n')
```
