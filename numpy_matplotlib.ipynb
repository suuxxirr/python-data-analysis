```python
import numpy as np
import matplotlib.pyplot as plt

# 1부터 5까지의 정수 중 랜덤하게 20개 생성
random_numbers = np.random.randint(1, 6, size = 20)

# 히스토그램 그리기
plt.hist(random_numbers, bins =range(1, 6), edgecolor='black', alpha=0.7)
plt.style.use('ggplot')
plt.rc('font', family = 'NanumBarunGothic')
plt.title('빈도수 히스토그램')
plt.xlabel('값')
plt.ylabel('빈도')
plt.xticks(range(1, 6))
plt.show()
```

- `bins=range(1,6)`: 가로축 구간 지정  
- `alpha`: 투명도
- `plt.style.use('ggplot')`: 격자 모양으로 출력

# 비율 원 그리는 프로그램 만들기

```python
import matplotlib.pyplot as plt

# 카테고리와 해당 카테고리의 비율
categories = ['A', 'B', 'C', 'D', 'E']
ratios = [25, 20, 15, 10, 30] # 각각의 비율, 전체 합은 100

# 원 그래프 그리기
plt.figure(figsize=(4, 4)) # 그림 크기(가로, 세로), 인치 단위
plt.pie(ratios, labels=categories, autopct='%1.1f%%', startangle=90, counterclock=False)
plt.title('카테고리 비율')
plt.axis('equal') # 원형을 유지하기 위해 가로세로 비율을 동일하게 설정
plt.show()
```

- `autopct='%1.1f%%'`: **autopercent**. 내부에 비율 표시. 소수점 한자리까지 표시, % 붙이기
- `startangle`: 원 그래프의 첫 데이터 시작 위치를 설정. 90 = 첫 데이터 시작을 12시 기준으로 설정
- `counterclock`: False는 출력 시계방향, True는 출력 시계반대방향
